Import Statements

```python
import pandas as pd
import numpy as np
```

Importing CSV files and inital dataframe creation

```python
student_filepath = 'PyCitySchools/Resources/students_complete.csv'
school_filepath = 'PyCitySchools/Resources/schools_complete.csv'
student_df = pd.read_csv(student_filepath)
school_df = pd.read_csv(school_filepath)
```

Merging Dataframes to consolidate the data

```python
combined_df = pd.merge(student_df, school_df, on='school_name', how='left')
combined_df
```

Creating the Summary Dataframe by consolidating and tabluating the contents of the previous merged dataframe

```python
total_schools = len(combined_df['school_name'].unique())
total_students = len(combined_df['Student ID'].unique())
total_budget = combined_df['budget'].unique().sum()
avg_math = combined_df['math_score'].mean()
avg_reading = combined_df['reading_score'].mean()
passing_math = len((combined_df[combined_df['math_score'] >= 70]))/(total_students) * 100
passing_reading = len((combined_df[combined_df['reading_score'] >= 70]))/(total_students) * 100
passing = len((combined_df.loc[(combined_df['reading_score'] >= 70) & (combined_df['math_score'] >= 70)]))/(total_students) * 100
summary_info = {'Total Schools': total_schools, 'Total Students': total_students, 'Total Budget': total_budget,
               'Average Math Score': avg_math, 'Average Reading Score': avg_reading, '% Passing Math': passing_math,
               '% Passing Reading': passing_reading, '% Overall Passing': passing}
summary_df = pd.DataFrame(summary_info, index=[0])
summary_df
```

Creating new dataframe containing the raw values desired for all future dataframe creation

```python
aggregation_function = {'type': 'first', 'student_name': 'count', 'budget': 'first', 'math_score': 'mean',
                        'reading_score': 'mean'}
df_new = combined_df.groupby('school_name').aggregate(aggregation_function)
school_list = df_new.index
math_vals = []
reading_vals= []
passing_vals= []
for school in school_list:
    iter_df = combined_df[combined_df['school_name'] == school]
    math = len((iter_df[iter_df['math_score'] >= 70]))/(len(iter_df)) * 100
    reading = len((iter_df[iter_df['reading_score'] >= 70]))/(len(iter_df)) * 100
    passing = len((iter_df.loc[(iter_df['reading_score'] >= 70) & (iter_df['math_score'] >= 70)]))/(len(iter_df)) * 100
    math_vals.append(math)
    reading_vals.append(reading)
    passing_vals.append(passing)
df_new['% Passing Math'] = math_vals
df_new['% Passing Reading'] = reading_vals
df_new['% Overall Passing'] = passing_vals
df_new['Budget per Student'] = (df_new['budget'])/(df_new['student_name'])
df_new = df_new.rename(columns={'type': 'School Type', 'student_name': 'Total Students', 'budget': 'Total School Budget',
                               'math_score': 'Average Math Score', 'reading_score': 'Average Reading Score'})
df_new = df_new[['School Type', 'Total Students', 'Total School Budget','Budget per Student', 'Average Math Score', 
                'Average Reading Score', '% Passing Math','% Passing Reading', '% Overall Passing']]
df_new
```

Taking the master dataframe formatting it and making it display the desired values

```python
df_by_school = df_new.copy(deep=True)
df_by_school['Total Students'] = df_by_school['Total Students'].map("{:,}".format)
df_by_school['Total School Budget'] = df_by_school['Total School Budget'].map("${:,.2f}".format)
df_by_school['Budget per Student'] = df_by_school['Budget per Student'].astype('int')
df_by_school['Average Math Score'] = df_by_school['Average Math Score'].map("{:,.2f}".format)
df_by_school['Average Reading Score'] = df_by_school['Average Reading Score'].map("{:,.2f}".format)
df_by_school['% Passing Math'] = df_by_school['% Passing Math'].map("{:,.2f}%".format)
df_by_school['% Passing Reading'] = df_by_school['% Passing Reading'].map("{:,.2f}%".format)
df_by_school['% Overall Passing'] = df_by_school['% Overall Passing'].map("{:,.2f}%".format)
df_by_school
```

Getting the Top 5 highest pass rate schools

```python
highest_pct = df_new.sort_values(by=['% Overall Passing'], ascending=False)
highest_pct = highest_pct.head()
highest_pct['Total Students'] = highest_pct['Total Students'].map("{:,}".format)
highest_pct['Total School Budget'] = highest_pct['Total School Budget'].map("${:,.2f}".format)
highest_pct['Budget per Student'] = highest_pct['Budget per Student'].astype('int')
highest_pct['Average Math Score'] = highest_pct['Average Math Score'].map("{:,.2f}".format)
highest_pct['Average Reading Score'] = highest_pct['Average Reading Score'].map("{:,.2f}".format)
highest_pct['% Passing Math'] = highest_pct['% Passing Math'].map("{:,.2f}%".format)
highest_pct['% Passing Reading'] = highest_pct['% Passing Reading'].map("{:,.2f}%".format)
highest_pct['% Overall Passing'] = highest_pct['% Overall Passing'].map("{:,.2f}%".format)
highest_pct
```

Getting the Top 5 lowest pass rate schools

```python
lowest_pct = df_new.sort_values(by=['% Overall Passing'])
lowest_pct = lowest_pct.head()
lowest_pct['Total Students'] = lowest_pct['Total Students'].map("{:,}".format)
lowest_pct['Total School Budget'] = lowest_pct['Total School Budget'].map("${:,.2f}".format)
lowest_pct['Budget per Student'] = lowest_pct['Budget per Student'].astype('int')
lowest_pct['Average Math Score'] = lowest_pct['Average Math Score'].map("{:,.2f}".format)
lowest_pct['Average Reading Score'] = lowest_pct['Average Reading Score'].map("{:,.2f}".format)
lowest_pct['% Passing Math'] = lowest_pct['% Passing Math'].map("{:,.2f}%".format)
lowest_pct['% Passing Reading'] = lowest_pct['% Passing Reading'].map("{:,.2f}%".format)
lowest_pct['% Overall Passing'] = lowest_pct['% Overall Passing'].map("{:,.2f}%".format)
lowest_pct
```

Taking the overall dataframe and creating a series of groupby objects for each grade and merging them into a larger dataframe

```python
score_df = combined_df[['school_name', 'math_score', 'reading_score', 'grade']]
nine_df = score_df[score_df['grade'] == '9th']
ten_df = score_df[score_df['grade'] == '10th']
eleven_df = score_df[score_df['grade'] == '11th']
twelve_df = score_df[score_df['grade'] == '12th']
nine_group = nine_df.groupby(['school_name'])
ten_group = ten_df.groupby(['school_name'])
eleven_group = eleven_df.groupby('school_name')
twelve_group = twelve_df.groupby('school_name')
nine_mean = nine_group.mean()
ten_mean = ten_group.mean()
eleven_mean = eleven_group.mean()
twelve_mean = twelve_group.mean()
nine_mean = nine_mean.rename(columns={'math_score': '9th', 'reading_score': '9th R'})
ten_mean = ten_mean.rename(columns= {'math_score': '10th', 'reading_score': '10th R'})
eleven_mean = eleven_mean.rename(columns={'math_score': '11th', 'reading_score': '11th R'})
twelve_mean = twelve_mean.rename(columns={'math_score': '12th', 'reading_score': '12th R'})
combined_score_df = nine_mean.merge(ten_mean, on='school_name')
combined_score_df = combined_score_df.merge(eleven_mean, on='school_name')
combined_score_df = combined_score_df.merge(twelve_mean, on='school_name')
```

Taking that larger dataframe and paring down only the math scores by grade

```python
math_score_df = combined_score_df[['9th', '10th', '11th', '12th']]
math_score_df = math_score_df.rename_axis(None, axis=0)
math_score_df['9th'] = math_score_df['9th'].map('{:,.2f}'.format)
math_score_df['10th'] = math_score_df['10th'].map('{:,.2f}'.format)
math_score_df['11th'] = math_score_df['11th'].map('{:,.2f}'.format)
math_score_df['12th'] = math_score_df['12th'].map('{:,.2f}'.format)
math_score_df
```

Taking that larger dataframe and paring down only the reading scores by grade

```python
reading_score_df = combined_score_df[['9th R', '10th R', '11th R', '12th R']]
reading_score_df = reading_score_df.rename(columns={'9th R': '9th', '10th R': '10th', '11th R': '11th', '12th R': '12th'})
reading_score_df = reading_score_df.rename_axis(None, axis=0)
reading_score_df['9th'] = reading_score_df['9th'].map('{:,.2f}'.format)
reading_score_df['10th'] = reading_score_df['10th'].map('{:,.2f}'.format)
reading_score_df['11th'] = reading_score_df['11th'].map('{:,.2f}'.format)
reading_score_df['12th'] = reading_score_df['12th'].map('{:,.2f}'.format)
reading_score_df
```

Creating a dataframe that can create the budget range column to describe which bin a budget falls within

```python
bins = np.arange(575, 675, 20)
labels = ['575 to 595','595 to 615','615 to 635','635 to 655']
df_spending = df_new[['Average Math Score','Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing', 'Budget per Student']]
df_spending['Budget Range'] = pd.cut(df_new['Budget per Student'], bins, labels=labels)
```

Grouping by the aforementioned budget range column and formatting the values therin

```python
df_spending_group = df_spending.groupby('Budget Range')
df_spending_answer = df_spending_group.mean()
df_spending_answer = df_spending_answer[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]
for value in df_spending_answer.columns:
    df_spending_answer[value] = df_spending_answer[value].map("{:,.2f}".format)
df_spending_answer
```

# It is pretty clear that there is a trend of student spending being inversely proporational to average scores and overall pass rates

Binning, classifying and rearranging the dataframe based on the size of the school

```python
bins = [0, 1000, 2500, 5000]
labels = ['small', 'medium', 'large']
df_size = df_spending[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]
df_size['Total Students'] = df_new['Total Students']
df_size['School Size'] = pd.cut(df_size['Total Students'], bins, labels=labels)
df_size_group = df_size.groupby('School Size')
df_size_answer = df_size_group.mean()
df_size_answer = df_size_answer[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]
for value in df_size_answer.columns:
    df_size_answer[value] = df_size_answer[value].map("{:,.2f}".format)
df_size_answer
```

No need to bin to do a similar groupby for school type as that is already and existing value found in the original dataframe

```python
df_type = df_spending[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]
df_type['School Type'] = df_new['School Type']
df_type_group = df_type.groupby('School Type')
df_type_answer = df_type_group.mean()
for value in df_type_answer.columns:
    df_type_answer[value] = df_type_answer[value].map("{:,.2f}".format)
df_type_answer
```
